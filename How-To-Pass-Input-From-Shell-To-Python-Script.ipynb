<h3 align="center">How To Pass Input From Shell To Python Script?</h3>

##### **Write python script to get from shell**

```python
from datetime import datetime
import sys

tme = datetime.today().strftime('%b %e')+' 09'
for i in sys.stdin:
    if tme in i:
        print(i.strip())
```
